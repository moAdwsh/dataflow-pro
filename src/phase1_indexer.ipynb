```python
import random

def generate_transactions(n=10000):
    data = []
    for i in range(n):
        txn = {
            "id": random.randint(1000, 999999),
            "amount": random.randint(100, 10000)
        }
        data.append(txn)
    return data
```

```python
#bubble sort
def bubble_sort(data):
    arr = data.copy()
    n = len(arr)

    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j]["id"] > arr[j + 1]["id"]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr
```

```python
#insertion sort
def insertion_sort(data):
    arr = data.copy()

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j]["id"] > key["id"]:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr
```

```python
#selection sort
def selection_sort(data):
    arr = data.copy()
    n = len(arr)

    for i in range(n):
        min_idx = i

        for j in range(i + 1, n):
            if arr[j]["id"] < arr[min_idx]["id"]:
                min_idx = j

        arr[i], arr[min_idx] = arr[min_idx], arr[i]

    return arr
```

```python
#merge sort
def merge_sort(data):
    if len(data) <= 1:
        return data

    mid = len(data) // 2
    left = merge_sort(data[:mid])
    right = merge_sort(data[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i]["id"] < right[j]["id"]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result
```

```python
#quick sort
def quick_sort(data):
    if len(data) <= 1:
        return data

    pivot = data[0]
    less = []
    greater = []

    for item in data[1:]:
        if item["id"] <= pivot["id"]:
            less.append(item)
        else:
            greater.append(item)

    return quick_sort(less) + [pivot] + quick_sort(greater)
```

```python
#timesort benchmark
import time

def benchmark():
    data = generate_transactions(5000)

    for func in [bubble_sort, insertion_sort, selection_sort, merge_sort, quick_sort]:
        start = time.time()
        func(data)
        end = time.time()
        print(f"{func.__name__}: {end - start:.4f} sec")

    # Python built-in sort
    start = time.time()
    sorted(data, key=lambda x: x["id"])
    end = time.time()
    print(f"python sort(): {end - start:.4f} sec")

benchmark()
```

```python
#linear search
def linear_search(data, target_id):
    for item in data:
        if item["id"] == target_id:
            return item
    return None
```

```python
#binary search
def binary_search(data, target_id):
    left = 0
    right = len(data) - 1

    while left <= right:
        mid = (left + right) // 2

        if data[mid]["id"] == target_id:
            return data[mid]

        elif data[mid]["id"] < target_id:
            left = mid + 1
        else:
            right = mid - 1

    return None
```

```python
#bisect 
import bisect

def get_q3_sales(sorted_ids):
    # مثال: IDs بتمثل time-series
    start = bisect.bisect_left(sorted_ids, 300000)
    end = bisect.bisect_right(sorted_ids, 600000)

    return sorted_ids[start:end]
```
