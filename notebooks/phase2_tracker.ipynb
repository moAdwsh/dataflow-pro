```python
#singly linked list 
class StepNode:
    def __init__(self, step_name):
        self.step = step_name
        self.next = None
# Tracker
class AppliedStepsTracker:
    def __init__(self):
        self.head = None

    def add_step(self, step_name):
        new_node = StepNode(step_name)

        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

        print(f"[Tracker] Added step: {step_name}")
    def show_steps(self):
        current = self.head

        print("\nApplied Steps:")
        while current:
            print(f"- {current.step}")
            current = current.next
```

```python
#doubly linked list
class DoublyStepNode:
    def __init__(self, step_name):
        self.step = step_name
        self.next = None
        self.prev = None

#tracker 
class AdvancedTracker:
    def __init__(self):
        self.head = None
        self.current = None

    def add_step(self, step_name):
        new_node = DoublyStepNode(step_name)

        if not self.head:
            self.head = new_node
            self.current = new_node
        else:
            self.current.next = new_node
            new_node.prev = self.current
            self.current = new_node

        print(f"[Tracker] Added: {step_name}")

#undo
    def undo(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f"[Undo] Current step: {self.current.step}")
        else:
            print("[Undo] No more steps to undo")

#redo
    def redo(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f"[Redo] Current step: {self.current.step}")
        else:
            print("[Redo] No more steps to redo")

#current
    def show_current(self):
        if self.current:
            print(f"Current Step: {self.current.step}")
        else:
            print("No steps yet")
```
